```python
import numpy as np
import scipy.optimize as sp
from astropy.timeseries import LombScargle
import matplotlib.pyplot as plt
import pandas as pd
from astropy.time import Time
import astropy.units as u
```

# Cleaning the Data Set

```python
df = pd.read_csv('Master_Data.csv')
df
```

```python
time = df['JD_UTC']
flux = df['av_rel_flux T1']
error = df['std_T1']
comp = df["av_rel_flux C2"]

plt.scatter(time, flux)
#plt.xlim(2.45e6,2.46e6)
plt.xlabel('Time')
plt.ylabel('Flux')
```

```python
# CLEANED DATA SET

dropped = df.drop([30,63]) # use this one
dropped.where(df['JD_UTC'] < 2.4e6).dropna()

t_astro = dropped['JD_UTC'] * u.day # time
mags = dropped['av_rel_flux T1'] - np.mean(dropped['av_rel_flux T1']) * u.mag # signal
d_error = dropped['std_T1']
d_comp = dropped["av_rel_flux C2"]

plt.scatter(t_astro, mags)
plt.xlabel('Time')
plt.ylabel('Normalized Flux')
```

# Lomb-Scargle Periodogram w/ All Data

```python
ls, power = LombScargle(t_astro, mags, d_error, nterms = 2).autopower()
```

```python
plt.figure(figsize = (12, 6))

t_end = 8

plt.xlim(1, t_end)
plt.ylim(0, 1)
plt.plot(1/ls, power)
plt.xticks(np.arange(1, t_end, 0.2))

plt.xlabel("Period (Days)")
plt.ylabel("Lomb Scargle Power")
plt.grid()
```

```python
cyg_dict = {"Period": 1/ls, "Power": power}

cyg_df = pd.DataFrame(cyg_dict)
cyg_df.where(cyg_df["Period"] <= 8).dropna().sort_values(by = "Power", ascending = False).head(10)
```
